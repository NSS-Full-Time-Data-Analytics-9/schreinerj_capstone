#### FM3 - Average Number of Own Children Under 18 Per Family, by Type of Family: 1955 to present
Source:  U.S. Census Bureau, Current Population Survey, March and Annual Social and Economic Supplements, 2022 and earlier.

#### would be interesting to see avg # of children going back further

```python
fm3 = pd.read_excel('../data/single_parent/census/historical_family_tables/fm3.xls', skiprows=7, nrows=61)
pd.set_option('display.max_rows', None)
fm3.head(2)
```

```python
#drop empty columns or columns not needed 
fm3 = fm3.drop(['Total', 'Married couple', 'Man, no spouse present', 'Woman, no spouse present', 'Unnamed: 5', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12'], axis=1)

#drop empty rows, drop original rows and keep revised
fm3 = fm3.drop(labels=[0, 1, 2, 5, 16]).reset_index(drop=True)

#rename columns 
fm3 = fm3.rename(columns = {'Unnamed: 0': 'year', 'Total.1': 'avg_num_child_own_child_und18', 'Married couple.1': 'avg_num_child_married_couple', 'Man, no spouse present.1': 'avg_num_child_father_only', 'Woman, no spouse present.1': 'avg_num_child_mother_only'})

fm3                     
```

```python
fm3.info()
```

```python
fm3 = fm3.astype({'year': 'str'})
#locate additional letters in year column
fm3_avg_num_child = fm3[fm3['year'].str.match('\d{4}[a-zA-Z]?', na=False)]
#extract only the year
fm3_avg_num_child['year'] = fm3_avg_num_child['year'].str.extract('(\d{4})', expand=False).astype(int)
#sort by descending
fm3_avg_num_child = fm3_avg_num_child.sort_values('year', ascending=False)

fm3_avg_num_child
```
