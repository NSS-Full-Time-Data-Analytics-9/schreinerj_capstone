```python
states_pop = pd.read_excel('../data/single_parent/census/state_population_tables/state_regional_population_totals_2010_2019.xlsx', skiprows=3, nrows=56)
pd.set_option('display.max_rows', None)
states_pop.tail(3)
```

```python
states_pop['Unnamed: 0'] = states_pop['Unnamed: 0'].str.replace(re.escape('.'), '')
```

```python
#drop national, regional rows
states_pop = states_pop.drop(labels=[0, 1, 2, 3, 4]).reset_index(drop=True)
states_pop.head(3)
```

```python
#rename columns
states_pop.rename(columns = {'Unnamed: 0': 'name'})
```

states_pop.to_csv('../data/single_parent/census/state_population_tables/states.pop.csv', index = False)
